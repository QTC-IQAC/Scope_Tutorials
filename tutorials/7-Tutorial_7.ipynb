## 1.1) Create Systems: Water, Benzene, Formaldehyde

In Tutorial 1, we created a System called "water", and imported a source, which was an .xyz file of a water geometry.

Here, we will load this System and complement it with 3 more structures, and we will also create two more Systems: benzene and formaldehyde

````python
#from scope.classes_system import System
#from scope.read_write import load_binary

water_sys   = scope.load_binary(f"{tutorial_folder}2-Systems/water/water.npy")  ## This is to load a system from the binary file
benz_sys    = scope.System("benzene")                                         ## This is to create a new system from scratch
form_sys    = scope.System("formaldehyde")
````

````python
## We set the paths for the two new systems. For water, we did it in Tutorial 1. Have a look there for more options and information
list_of_systems = [water_sys]
for tsys in [benz_sys, form_sys]:
    tsys.sources_path       = f"{tutorial_folder}1-Sources/{tsys.name}/"
    tsys.system_path        = f"{tutorial_folder}2-Systems/{tsys.name}/"
    tsys.system_file        = f"{tutorial_folder}2-Systems/{tsys.name}/{tsys.name}.npy"
    tsys.computations_path  = f"{tutorial_folder}3-Computations/{tsys.name}/"
    list_of_systems.append(tsys)
````

````python
## We stored the 3 systems in a list, with very little information, only the paths, and a single source for water. 
print(list_of_systems)
````

## 1.2) Import Molecules in Sources Folder (sys.sources_path) 

````python
# Currently, SCOPE is designed so each system has its own folder with sources. In this example, the sources are .xyz files

# Notice that there is no universal way of creating systems. Each problem in computational chemistry faces its own system structure. 
# In the SCO add-on, the original data is in a cell2mol CELL format, so a new function had to be created to import molecules from there. 
# New add-ons are being prepared, each one with its own form of importing systems, and arranging molecules within those systems.

# For the moment, the simplest way to import sources is when those are in an XYZ format.
for tsys in list_of_systems: 
    tsys.load_multiple_xyz(tsys.sources_path)
    tsys.save()
````

````python
## When you run the cell above for the first time, you should receive a WARNING telling you that a source with the name "geom1" exists for water. 
## This is because Systems do not accept new sources whose name is already used by an existing source. This would be dangerous. 
## Remember in Tutorial 1 we already imported geom1.xyz

## Similarly, if you try to re-run the cell above, you'll receive even more warnings 
````

````python
## We now added the .xyz files with molecules as sources. Our systems are ready:
print(list_of_systems)
````

## 2) Create SCOPE input files for tasks 2 and 3

In tutorial 5, we created the input for task1, which was a Single Point computation. Now we'll do the same for task2 and task3:

- Task2 will be a geometry optimization.
- Task3 will be a frequency computation. 

Notice that, as opposed to Gaussian16, where you can specify both tasks in a single input file, in **SCOPE** these belong to different tasks. 

````python
## See Tutorial 5 for more information

task1_input = """
&environment
   max_jobs         = 24
   max_procs        = 96
   requested_procs  = 4
/

&options
/

&job_data
   branch       = 'tutorial'
   workflow     = 'all'
   job_setup    = 'regular'
   suffix       = 'opt'
   keyword      = 'pbe opt'

   istate       = 'initial'
   fstate       = 'pbe_opt'

   hierarchy    = 2
   requisites   = ['pbe scf']
   constrains   = ['self']
   must_be_good = True
/

&qc_data
   software     = 'g16'
   jobtype      = 'opt'
   functional   = 'pbe'
   basis        = 'sto-3g'
   is_grimme    = True
   loose_opt    = True
/"""
````

````python
task2_input = """
&environment
   max_jobs         = 24
   max_procs        = 96
   requested_procs  = 4
/

&options
/

&job_data
   branch       = 'tutorial'
   workflow     = 'all'
   job_setup    = 'regular'
   suffix       = 'freq'
   keyword      = 'pbe freq'

   istate       = 'pbe_opt'
   fstate       = 'pbe_opt'

   hierarchy    = 3
   requisites   = ['pbe opt']
   constrains   = ['self']
   must_be_good = True
/

&qc_data
   software     = 'g16'
   jobtype      = 'freq'
   functional   = 'pbe'
   basis        = 'sto-3g'
   is_grimme    = True
/"""
````

````python
# IMPORTANT. Notice that, when running the 'scope_run_job' script. The job file (-j arg) must be an actual file, not a string as in the cell above
# Thus, we save this text to an actual file, so we can call it later
file_name1 = ''.join([tutorial_folder,"task2.scope"])
file_name2 = ''.join([tutorial_folder,"task3.scope"])
scope.read_write.save_text(task1_input, file_name1)
scope.read_write.save_text(task2_input, file_name2)
````

## 3) Run SCOPE

Once the system and a job_file are prepared, we can submit the computation in the appropriate cluster. To do so:

**(1) Configure** the Environment in the cluster by typing "scope config":

- The computation cluster must have Gaussian 16 installed
- You can give the environment any name, but for the sake of simplicity, here we will call it "scope_env_cluster.npy" 

**(2) Upload** the system and scope input files to the cluster. In principle, you should upload the system files to dedicated folders inside the **env.systems_path** of the cluster enviromnent (env).  

For instance, if env.sys_path = '/home/user/Scope_Tutorials/Data/7-Tutorial_7/2-Systems', the system file should go to: 
- '/home/user/Scope_Tutorials/Data/7-Tutorial_7/2-Systems/water/water.npy'
- '/home/user/Scope_Tutorials/Data/7-Tutorial_7/2-Systems/benzene/benzene.npy'
- '/home/user/Scope_Tutorials/Data/7-Tutorial_7/2-Systems/formaldehyde/formaldehyde.npy'

In practice, you can upload the system files anywhere in the cluster. However, the cluster's Environment will force the systems to be saved in their expected paths above. This is the minor price to pay for the Environment controlling the paths.    

**(3) Run**: 'scope run'. If the paths are the expected ones, you should type: 

```bash
scope run -n scope_env_cluster.npy -s 2-Systems/water/water.npy -t task1.scope task2.scope task3.scope -e 
scope run -n scope_env_cluster.npy -s 2-Systems/benzene/benzene.npy -t task1.scope task2.scope task3.scope -e 
scope run -n scope_env_cluster.npy -s 2-Systems/formaldehyde/formaldehyde.npy -t task1.scope task2.scope task3.scope -e
```

The <span style="color:blue">**-e**</span> option handles some errors that are relatively common with Gaussian16 or Quantum Espresso, at least in the computational clusters where I could test SCOPE. These 'software' errors could not be solved by the support teams of those clusters, so they seem to be inherent. They appear randomly, and are solved by re-submitting the job. Thus, whenever one of those errors is detected, SCOPE will just delete the aborted output and resubmit. 

**(4) Check and Wait**. 

- If everything worked well, then you should have gaussian16 jobs running in your cluster. Whenever one -or more- of those jobs finish, proceed to the next tutorial
- Otherwise, if all jobs fail or are aborted by SLURM/SGE, check the 'gen_g16_subfile' function in 'core/src/scope/software/g16_input.py'.  
  This function creates the .sub files for G16 jobs. You might have to change it depending on how the computation cluster is set up.   
  If you fail to make it work, please contact me at sergi.vela@iqac.csic.es
